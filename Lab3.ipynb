# Σημειώσεις

## 1

Ένα από τα πρώτα σημαντικά αποτελέσματα στην θεωρία της κβαντικής πληροφορίας ήταν η απόδειξη ότι δεν μπορούμε να κατασκευάσουμε κάποιο κύκλωμα που να αντιγράφει οποιαδήποτε κατάσταση ενός qubit. Ας υποθέσουμε προς το παρών ότι κάτι τέτοιο είναι δυνατόν και ότι υπάρχει κάποιο κύκλωμα που αναπαριστάται από κάποιον μοναδιαίο μετασχηματισμό $U$ έτσι ώστε η δράση του πάνω στο qubit που θέλουμε να αντιγράψουμε και σε ένα ακόμη qubit προετοιμασμένο σε μία αρχική κατάσταση $\left|s\right>$ να είναι η εξής

$$U(\left|\psi\right>\left|s\right>)=\left|\psi\right>\left|\psi\right>$$

Έστω τώρα ότι αντιγράφουμε ένα qubit προετοιμασμένο σε μία διαφορετική κατάσταση $\left|\phi\right>$, όπως προηγουμένως ισχύει

$$U(\left|\phi\right>\left|s\right>)=\left|\phi\right>\left|\phi\right>$$

Παίρνοντας το εσωτερικό γινόμενο αυτών των δύο εξισώσεων καταλήγουμε στην σχέση

$$\left<\psi|\phi\right>=\left<\psi|\phi\right>^2$$

Η οποία μπορεί να ισχύει μονάχα αν $\left<\psi|\phi\right>=0$ ή $\left<\psi|\phi\right>=1$, το κύκλωμά μας δηλαδή μπορεί να αντιγράψει μονάχα καταστάσεις οι οποίες είναι κάθετες μεταξύ τους όπως συμβαίνει με την βάση $\{\left|0\right>,\left|1\right>\}$.

## 2
Στο πραγματικό πρωτόκωλλο τηλεμεταφοράς όπου στόχος είναι η μεταφορά της κατάστασης σε κάποιο άλλο σημείο οσοδήποτε μακριά, οι μετρήσεις στα δύο πρώτα qubit πραγματοποιούνται αμέσως μετά την εφαρμογή της πύλης Hadamard. Ανάλογα με το αποτέλεσμα αυτής της μέτρησης θα πρέπει να δράσουμε στο τρίτο qubit με την πύλη που φαίνεται στον παρακάτω πίνακα

\begin{array}{cc} 
Αποτέλεσμα & Δράση\\ \hline
00 & I \\
01 & X \\
10 & Z \\
11 & ZX
\end{array}

Όπως είδαμε όμως στην Άσκηση 3.2 μπορούμε πάντα να αφήσουμε την διαδικασία της μέτρησης για το τέλος αρκεί να αντικαταστήσουμε όπου χρειάζεται τις κατάλληλες controlled πύλες.

## 3

Δοσμένης χρονοανεξάρτητης Χαμιλτονιανής $H$ η εξέλιξη ενός κβαντικού συστήματος μετά από χρόνο $t$ δίνεται από την δράση του μοναδιαίου τελεστή $U(t)=e^{-itH}$ (σε σύστημα μονάδων όπου $\hbar=1$) πάνω στην αρχική κατάσταση

$$\left|\psi(t)\right>=e^{-itH}\left|\psi(t=0)\right>$$

Στην ειδική περίπτωση δε όπου η Χαμιλτονιανή είναι το άθροισμα $Η=\sum_{i=1}^n\sigma_i$ όπου $\sigma_i$ οι πίνακες $\{I,X,Y,Z\}$

$$e^{-itH}=\otimes_{i=1}^n e^{-it\sigma_i}$$

Παρατηρώντας τώρα ότι

\begin{align}
        e^{-itX}&=R_x(2t)=U_3(2t,-\pi,\pi/2)\\
        e^{-itY}&=R_y(2t)=U_3(2t,0,0)\\
        e^{-itZ}&=R_z(2t)=e^{it}U_1(2t)
\end{align}

Aρκεί στο κύκλωμά μας να τοποθετήσουμε την κατάλληλη κάθε φορά πύλη για να προσομοιώσουμε την χρονική εξέλιξη του συστήματος.

Εαν η Χαμιλτονιανή έχει την μορφή $Η=\otimes_{i=1}^n\sigma_i$ τότε παρατηρώντας ότι $U(t)=\cos(t)-i\sin(t)\otimes_{i=1}^n\sigma_i$ κατασκευάζουμε το ακόλουθο κύκλωμα (δείχνεται η περίπτωση όπου n=2 και $t=1$ η οποία γενικεύεται άμεσα).

![a](img/lab3/circuit3.4.png)

To κύκλωμα αυτό προσομοιάζει μία Χαμιλτονιανή της μορφής $H=Ζ\otimes Z$ και άρα $U(t)=\cos(t)-i\sin(t)Z\otimes Z$ (στην γενική περίπτωση το γινόμενο περιέχει $n$ τέτοιους όρους). Για να προσομοιώσουμε μία Χαμιλτονιανή της μορφής $\sigma_1\otimes\sigma_2$ αρκεί να έχουμε υπόψιν μας τις ακόλουθες σχέσεις

\begin{align}
    X&=U_3(\pi/2,0,0)ZU_3(-\pi/2,0,0)\\
    Y&=U_3(-\pi/2,-\pi,\pi/2)ZU_3(\pi/2,-\pi,\pi/2)
\end{align}

επομένως θα πρέπει να δράσουμε στην αρχή και στο τέλος με την κατάλληλη πύλη $U_3$. Για παράδειγμα αν θέλουμε να προσομοιώσουμε την Χαμιλτονιανή $H=X\otimes Y$ για χρόνο $t=2$ μετατρέπουμε το παραπάνω κύκλωμα ώς εξής
